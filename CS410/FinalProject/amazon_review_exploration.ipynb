```python
import pandas as pd
import numpy as np
import gzip
```

```python
#Helper methods to parse raw review files
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield eval(l)
```

```python
def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')
```

```python
review_file = 'reviews_Electronics_5.json.gz'
df = getDF(review_file)
```
